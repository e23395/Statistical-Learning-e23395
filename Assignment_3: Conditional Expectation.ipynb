**2. Compute the $\sigma$-algebra generated by $Y$, namely $\sigma(Y)$**

The Random variable Y maps $\Omega$ to $\{0,1\}$. To find $\sigma(Y)$, we find the preimages of all possible subsets of the range $\{0,1\}$

*   $Y^{-1}(\emptyset) = \emptyset$
*   $Y^{-1}(\{0\}) = \{\omega_1,\omega_2\}$
*   $Y^{-1}(\{1\}) = \{\omega_3,\omega_4\}$

$$\sigma(Y) = \{\emptyset, \{\omega_1,\omega_2\},\{\omega_3,\omega_4\},\Omega\}$$

**3. Explain why $\mathbb{E}[X \mid Y]$ must be $\sigma(Y)$-measurable, and therefore must be constant on the sets $\{\omega_1, \omega_2\}$ and $\{\omega_3, \omega_4\}$**

By definition, the conditional expectation $\mathbb{E}[X|Y]$ is a random variable whose value depends only on the information provided by $Y$.
*   Because it relies entirely on the sub$- \sigma -$algebra $\sigma(Y)$, it must be $\sigma(Y)-$measurable.
*   The smallest building blocks of $\sigma(Y)$ are the sets $\{\omega_1,\omega_2\}$ and $\{\omega_3,\omega_4\}$.
*   Since a measurable function cannot split the atoms of its underlying $\sigma-$ algebra, $\mathbb{E}[X|Y]$ cannot distinguish between $\omega_1$ and $\omega_2$, nor between $\omega_3$ and $\omega_4$.Thus it must take a constant value on each of these two sets.
$$\mathbb{E}[X|Y] = \begin{cases} a,\omega \in \{\omega_1, \omega_2\} \\
b, \omega \in \{\omega_3, \omega_4\}
\end{cases}$$

for some constants $a$ and $b$ to be determined.

# **Q2) Conditional Density of the Multivariate Gaussian**

**Step 1: The Conditional Density Formula**
$$f_{\mathbf{X} \mid \mathbf{Y}}(\mathbf{x} \mid \mathbf{y}) = \frac{f_{\mathbf{X}, \mathbf{Y}}(\mathbf{x}, \mathbf{y})}{f_{\mathbf{Y}}(\mathbf{y})}$$

Since $\mathbf{Z} = \begin{bmatrix} \mathbf{X} \\ \mathbf{Y} \end{bmatrix}$ is jointly Gaussian, the marginal distribution of $\mathbf{Y}$ is simply:

$$\mathbf{Y} \sim \mathcal{N}(\boldsymbol{\mu}_Y, \boldsymbol{\Sigma}_{YY})$$

**Step 2: Block Matrix Inversion (Schur Complement)**
$$\boldsymbol{\Sigma}^{-1} =
 \begin{bmatrix} \boldsymbol{\Sigma}_{XX} & \boldsymbol{\Sigma}_{XY} \\
  \boldsymbol{\Sigma}_{YX} & \boldsymbol{\Sigma}_{YY}
  \end{bmatrix}^{-1} =
  \begin{bmatrix} \mathbf{S}_{XX}^{-1} & -\mathbf{S}_{XX}^{-1}\boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1} \\
  -\boldsymbol{\Sigma}_{YY}^{-1}\boldsymbol{\Sigma}_{YX}\mathbf{S}_{XX}^{-1} & \boldsymbol{\Sigma}_{YY}^{-1} + \boldsymbol{\Sigma}_{YY}^{-1}\boldsymbol{\Sigma}_{YX}\mathbf{S}_{XX}^{-1}\boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}
  \end{bmatrix}$$

  Where $\mathbf{S}_{XX}$ is the Schur complement of $\boldsymbol{\Sigma}_{YY}$ in $\boldsymbol{\Sigma}$:
  $$\mathbf{S}_{XX} = \boldsymbol{\Sigma}_{X \mid Y} = \boldsymbol{\Sigma}_{XX} - \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}\boldsymbol{\Sigma}_{YX}$$

**Step 3: Expanding the Quadratic Form in the Exponent**
$$\mathbf{Q} = (\mathbf{z} - \boldsymbol{\mu})^T \boldsymbol{\Sigma}^{-1} (\mathbf{z} - \boldsymbol{\mu})$$
Let $\tilde{\mathbf{x}} = \mathbf{x} - \boldsymbol{\mu}_X$ and $\tilde{\mathbf{y}} = \mathbf{y} - \boldsymbol{\mu}_Y$. Substituting the block matrix form of $\boldsymbol{\Sigma}^{-1}$:
$$\mathbf{Q} = \begin{bmatrix} \tilde{\mathbf{x}}^T & \tilde{\mathbf{y}}^T \end{bmatrix} \begin{bmatrix} \mathbf{S}_{XX}^{-1} & -\mathbf{S}_{XX}^{-1}\boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1} \\ -\boldsymbol{\Sigma}_{YY}^{-1}\boldsymbol{\Sigma}_{YX}\mathbf{S}_{XX}^{-1} & \boldsymbol{\Sigma}_{YY}^{-1} + \boldsymbol{\Sigma}_{YY}^{-1}\boldsymbol{\Sigma}_{YX}\mathbf{S}_{XX}^{-1}\boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1} \end{bmatrix} \begin{bmatrix} \tilde{\mathbf{x}} \\ \tilde{\mathbf{y}} \end{bmatrix}$$

$$\mathbf{Q} = \tilde{\mathbf{y}}^T \boldsymbol{\Sigma}_{YY}^{-1} \tilde{\mathbf{y}} + (\tilde{\mathbf{x}} - \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}\tilde{\mathbf{y}})^T \mathbf{S}_{XX}^{-1} (\tilde{\mathbf{x}} - \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}\tilde{\mathbf{y}})$$

**Step 4: Determinant Decomposition**
$$|\boldsymbol{\Sigma}| = |\boldsymbol{\Sigma}_{YY}| \cdot |\mathbf{S}_{XX}|$$

**Step 5: Combining into the Conditional Density**
$$f_{\mathbf{X} \mid \mathbf{Y}}(\mathbf{x} \mid \mathbf{y}) =
\frac{\frac{1}{\sqrt{(2\pi)^{n+m}|\boldsymbol{\Sigma}_{YY}||\mathbf{S}_{XX}|}} \exp\left(-\frac{1}{2}\left[\tilde{\mathbf{y}}^T \boldsymbol{\Sigma}_{YY}^{-1} \tilde{\mathbf{y}} + (\tilde{\mathbf{x}} - \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}\tilde{\mathbf{y}})^T \mathbf{S}_{XX}^{-1} (\tilde{\mathbf{x}} - \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}\tilde{\mathbf{y}})\right]\right)}{\frac{1}{\sqrt{(2\pi)^m|\boldsymbol{\Sigma}_{YY}|}} \exp\left(-\frac{1}{2}\tilde{\mathbf{y}}^T \boldsymbol{\Sigma}_{YY}^{-1} \tilde{\mathbf{y}}\right)}$$

The terms involving $\tilde{\mathbf{y}}^T \boldsymbol{\Sigma}_{YY}^{-1} \tilde{\mathbf{y}}$ and the factor $\sqrt{(2\pi)^m|\boldsymbol{\Sigma}_{YY}|}$ cancel out perfectly, leaving:

$$f_{\mathbf{X} \mid \mathbf{Y}}(\mathbf{x} \mid \mathbf{y}) = \frac{1}{\sqrt{(2\pi)^n|\mathbf{S}_{XX}|}} \exp\left(-\frac{1}{2}(\mathbf{x} - \boldsymbol{\mu}^*)^T \mathbf{S}_{XX}^{-1} (\mathbf{x} - \boldsymbol{\mu}^*)\right)$$

Where:
*   $\boldsymbol{\mu}^* = \boldsymbol{\mu}_X + \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}(\mathbf{y} - \boldsymbol{\mu}_Y)$
*   $\mathbf{S}_{XX} = \boldsymbol{\Sigma}_{X \mid Y} = \boldsymbol{\Sigma}_{XX} - \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}\boldsymbol{\Sigma}_{YX}$




#**Q3) Continuous Fault Monitoring of a Spindle Motor**

**1. Joint Distribution**<br>
Given: $X \sim \mathcal{N}(0,1)$, $Q \sim \mathcal{N}(0,4)$, where $X$ and $Q$ are independent, and $Y = X + Q$.
*   Mean: $\mu_Y = \mathbb{E}[X + Q] = \mathbb{E}[X] + \mathbb{E}[Q] = 0 + 0 = 0$
*   Variance: $\text{Var}(Y) = \text{Var}(X + Q) = \text{Var}(X) + \text{Var}(Q) = 1 + 4 = 5$
*   Thus, $Y \sim \mathcal{N}(0, 5)$.
*   Find the covariance $\text{Cov}(X,Y)$:
$$\text{Cov}(X, Y) = \text{Cov}(X, X + Q) = \text{Cov}(X,X) + \text{Cov}(X,Q)$$
Since $X$ and $Q$ are independent, $\text{Cov}(X,Q) = 0$.
$$\text{Cov}(X,Y) = \text{Var}(X) = 1$$

---

**2. Information Geometry**
*   Why $X$ is not $\sigma(Y)$-measurable:<br>
The sub-$\sigma$-algebra $\sigma(Y)$ contains only the information revealed by the sum $Y$. Because infinitely many combinations of $X$ and $Q$ can yield the exact same sum $Y$, you cannot uniquely map an observed value of $Y$ back to a unique value of $X$. Thus, $X$ is not a function of $Y$.

*   Practical Interpretation:<br>
If $Y = 5$, it could mean the motor is dangerously overheating ($X = 5, Q = 0$) or it could mean the temperature is perfectly fine but the machine is cutting an extremely hard material ($X = 0, Q = 5$). A single combined power reading cannot distinguish between thermal load and mechanical load.

---

**3. Conditional Expectation (The Best Estimate)**

*   Compute $\mathbb{E}[X \mid Y = y]$:<br>
$$\mathbb{E}[X \mid Y = y] = \mu_X + \frac{\text{Cov}(X,Y)}{\text{Var}(Y)}(y - \mu_Y) = 0 + \frac{1}{5}(y - 0) = \frac{1}{5}y$$
$$\mathbb{E}[X \mid Y] = 0.2Y$$

*   Signal-to-Total-Variance Ratio:
$$\mathbb{E}[X \mid Y] = \left(\frac{\text{Var}(X)}{\text{Var}(X) + \text{Var}(Q)}\right) Y = \left(\frac{1}{1 + 4}\right) Y = \frac{1}{5}Y$$

---

**4. Verification of the Residual** <br>
Let $\widehat{X} = \mathbb{E}[X \mid Y] = \frac{1}{5}Y$. The residual error is $Z = X - \widehat{X} = X - \frac{1}{5}Y$.

*   Prove $\mathbb{E}[Z] = 0$:
$$\mathbb{E}[Z] = \mathbb{E}\left[X - \frac{1}{5}Y\right] = \mathbb{E}[X] - \frac{1}{5}\mathbb{E}[Y] = 0 - \frac{1}{5}(0) = 0$$

*   Verify the Orthogonality Principle ($\text{Cov}(Z, Y) = 0$):
$$\text{Cov}(Z, Y) = \text{Cov}\left(X - \frac{1}{5}Y, Y\right) = \text{Cov}(X,Y) - \frac{1}{5}\text{Cov}(Y,Y)$$
$$\text{Cov}(Z,Y) = 1 - \frac{1}{5}\text{Var}(Y) = 1 - \frac{1}{5}(5) = 1 - 1 = 0$$

---

**5. Information Theory: Differential Entropy**<br>
For a normal distribution $\mathcal{N}(\mu, \sigma^2)$, the differential entropy is $h = \frac{1}{2}\ln(2\pi e \sigma^2)$.
*   Compute $h(X)$:
$$h(X) = \frac{1}{2}\ln(2\pi e \cdot 1) = \frac{1}{2}\ln(2\pi e)$$
*   Compute $h(X \mid Y)$:
$$\sigma^2_{X \mid Y} = \sigma^2_X - \frac{\text{Cov}(X,Y)^2}{\text{Var}(Y)} = 1 - \frac{1^2}{5} = \frac{4}{5} = 0.8$$
$$h(X \mid Y) = \frac{1}{2}\ln\left(2\pi e \cdot \frac{4}{5}\right) = \frac{1}{2}\ln(2\pi e) + \frac{1}{2}\ln\left(\frac{4}{5}\right)$$
---

**6. Mutual Information**
*   Calculate $I(X; Y)$:
$$I(X; Y) = h(X) - h(X \mid Y) = \frac{1}{2}\ln(2\pi e) - \left[\frac{1}{2}\ln(2\pi e) + \frac{1}{2}\ln\left(\frac{4}{5}\right)\right]$$
$$I(X;Y) = -\frac{1}{2}\ln\left(\frac{4}{5}\right) = \frac{1}{2}\ln\left(\frac{5}{4}\right) \approx 0.1116 \text{ nats}$$

*   Percentage of "uncertainty" removed:<br>
Using the entropy
$$\text{Percentage} = \frac{I(X;Y)}{h(X)} = \frac{\frac{1}{2}\ln(5/4)}{\frac{1}{2}\ln(2\pi e)} = \frac{\ln(1.25)}{\ln(2\pi e)} \approx \frac{0.2231}{2.8379} \times 100\% \approx 7.86\%$$

*   How this changes if $\text{Var}(Q) = 100$:<br>
If noise increases to $100$, then $\text{Var}(Y) = 1 + 100 = 101$
$$\sigma^2_{X \mid Y} = 1 - \frac{1^2}{101} = \frac{100}{101}$$
$$I(X;Y) = -\frac{1}{2}\ln\left(\frac{100}{101}\right) = \frac{1}{2}\ln\left(\frac{101}{100}\right) \approx 0.00497 \text{ nats}$$
$$\text{Percentage} = \frac{\ln(1.01)}{\ln(2\pi e)} \times 100\% \approx \frac{0.00995}{2.8379} \times 100\% \approx 0.35\%$$
The massive mechanical noise completely drowns out the temperature signal, meaning almost zero uncertainty is removed.

---
**7. Diagnostic Interpretation**
*   Why $Y$ is a biased estimator of $X$:<br>
An estimator $\widehat{X}$ is conditionally unbiased if $\mathbb{E}[\widehat{X} \mid X] = X$. If we set $\widehat{X} = Y = X + Q$, then:
$$\mathbb{E}[Y \mid X] = \mathbb{E}[X + Q \mid X] = X + \mathbb{E}[Q] = X$$
While it is unconditionally unbiased on average, using $Y$ directly to estimate a specific state overestimates the scale of fluctuations because it includes the variance of $Q$.

*   Why $\mathbb{E}[X \mid Y]$ "shrinks" the measurement:
Our optimal estimate is $\mathbb{E}[X \mid Y] = \frac{1}{5}Y$. The factor $\frac{1}{5}$ acts as a attenuation scaling factor. Because we know $Y$ is contaminated by load noise ($Q$), any large movement in $Y$ is statistically more likely to have been caused by a fluctuation in the high-variance load ($\text{Var}(Q)=4$) rather than the lower-variance temperature ($\text{Var}(X)=1$).<br>
To minimize the Mean Square Error, the mathematics pulls (shrinks) the raw reading $Y$ closer to the historical mean of zero, trusting it less as noise dominates.
