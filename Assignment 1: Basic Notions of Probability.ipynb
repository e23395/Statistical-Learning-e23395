# **Q1) The probability of shooting at a target.**





1.   **Uniformly distributed with respect to Area**


*   **Total area (S): The area of a disk with radius R = 10**
$$
Area_t = πR^2 = π(10)^2 = 100π
$$
*   **Event A(Within one unit of the boundary):** This is the outer ring of the disk. The boundry is at $r = 10$, so within one unit means the point lies between $r = 9$ and $r = 10$.


*   **Area of event A**:
$$
Area_A = π(10)^2 - π(9)^2 = 100π - 81π = 19π
$$
*   **Probability P(A)**:
$$
P(A)=\frac{\text{Area of}A}{\text{Area of disk}}
=\frac{19π}{100π}
=\frac{19}{100}
=0.19
$$




2.   **Uniformly Distributed with Respect to Distanceand Direction**


*   In this model, we choose a radius ($r$) and angle ($θ$) independently.
*   **Distance ($r$):** chosen uniformly from $[0,10]$
*   **Direction ($θ$):** chosen uniformly from $[0,2π)$
*   **Event ($A$):** The point lies within 1 unit of the boundary, which means  $[9\le r \le 10]$.
*   **Calculating P($A$):** since $r$ is chosen uniformly from a linear interval of length 10, the probability is simply the length of the favourable interval devided by total length.
$$
P(A) = \frac{10-9}{10-0}
=\frac{1}{10}
=0.10
$$



3.   **Why the two answers differ**


*   Model 1 (Area): To keep the distribution uniform across the area, there must be more points far from the center than near it. This is because there is "more room" (more area) at larger radii. As a result, $P(A)$ is higher $(0.19)$ because the outer ring occupies a significant portion of the total area.
*   Model 2 (Distance/Direction): This model treats every possible distance from the center as equally likely. It essentially "clumps" more points toward the center relative to the available area. Imagine drawing many circles at equal intervals; in this model, you are just as likely to pick a point on a tiny circle near the center as you are on the large boundary circle. This makes the outer boundary region seem "smaller" in terms of probability, leading to a lower value $(0.10)$.

# **Q2) Bertrand's Paradox**

The goal is to find the probability that a chord is longer than the side of an inscribed equilateral triangle ($L = R\sqrt{3}$).




**Method 1: Random Endpoints**


*   **Step 1:** Fix one endpoint of the chord at a vertex of the inscribed triangle.
*   **Step 2:** Observe that the chord is longer than the side $L$ only if the second endpoint lies on the arc between the other two vertices.


*   **Step 3:** Since the triangle is equilateral, this arc is exactly $1/3$ of the circle's circumference.
*   **Step 4:**
$$
P = \frac{1}{3}
$$


**Method 2: Random Radial Distancet**


*   **Step 1:** Choose a random radius and a random point along it to be the chord's midpoint.

*   **Step 2:** Geometrically, a chord is longer than $L$ if its midpoint is closer to the center than the midpoint of the triangle's side.
*   **Step 3:** The midpoint of the side of an inscribed equilateral triangle is at distance $R/2$ from the center.


*   **Step 4:** The favorable distance is the interval $[0, R/2]$ out of the total radius $[0, R]$.


*   **Step 5:**
$$
P = \frac{R/2}{R} = \frac{1}{2}
$$






**Method 3: Random Midpoint**


*   **Step 1:** Choose a random point anywhere inside the circle to be the midpoint of the chord.

*   **Step 2:** A chord is longer than $L$ if its midpoint falls within a smaller concentric circle of radius $R/2$.
*   **Step 3:** Calculate the ratio of the area of the inner circle to the total circle.
$$
P = \frac{\pi(R/2)^2}{\pi R^2} = \frac{1/4}{1} = \frac{1}{4}
$$

# **Q8) Probability of Being a Girl?**



1.   **Identify the Sample Space**:
For a family with two children, there are four equally likely possibilities (where the first letter is the first child and the second letter is the second child):

*   $BB (Boy, Boy)$
*   $BG (Boy, Girl)$
*   $GB (Girl, Boy)$
*   $GG (Girl, Girl)$

2.   **Apply the Condition (Reduced Sample Space)**: The problem states we are given that at least one child is a girl. This information eliminates the "BB" option from our possibilities.

*   New Sample Space = {$BG$, $GB$, $GG$}
*   Total number of possible outcomes $= 3$


3.   **Identify the Favorable Outcome** : We want the probability that both children are girls.


*   Favorable outcome $=$ {$GG$}
*   Number of favorable outcomes $= 1$


4.   **Calculate the Probability**
$$
P(\text{Both Girls} \mid \text{At least one Girl}) = \frac{\text{Number of favorable outcomes}}{\text{Total possible outcomes}} = \mathbf{\frac{1}{3}}
$$



# **Q9) Information Available in a Discrete Random Variable**



1.   **Probability Space $(\Omega, \mathcal{F}, P)$**

*   **Sample Space ($\Omega$):** The set of all possible outcomes.
$$\Omega = \{G_1, G_2, D_1, D_2\}$$
*   **$\sigma$-algebra ($\mathcal{F}$):** Since $\mathcal{F} = \mathfrak{P}(\Omega)$ (the power set), it contains all $2^4 = 16$ possible subsets of $\Omega$.
$$\mathcal{F} = \{\emptyset, \{G_1\}, \{G_2\}, \{D_1\}, \{D_2\}, \{G_1, G_2\}, \{G_1, D_1\}, \dots, \Omega\}$$
*   **Probability Measure ($P$):** Defined by the given point masses:
$$P(G_1)=0.50, P(G_2)=0.20, P(D_1)=0.10, P(D_2)=0.20$$
****


2.   **$\sigma$-algebra generated by $X$, $\sigma(X)$**

*   $X$ records quality: $X=0$ for $\{G_1, G_2\}$ and $X=1$ for $\{D_1, D_2\}$.
The $\sigma$-algebra $\sigma(X)$ is the collection of sets $\{X \in B\}$ for all Borel sets $B$.
*   $\sigma(X) = \{\emptyset, \{G_1, G_2\}, \{D_1, D_2\}, \Omega\}$
****



3.   **$\sigma$-algebra generated by $Y$, $\sigma(Y)$**
*   $Y$ records quality and machine: $Y=1$ for $\{G_1\}$, $Y=2$ for $\{G_2\}$, $Y=3$ for $\{D_1\}$, and $Y=4$ for $\{D_2\}$.
Since $Y$ assigns a unique value to every individual outcome in $\Omega$:
*   $\sigma(Y) = \mathfrak{P}(\Omega)$ (The power set calculated in step 1).
****

4.   **Comparison of $\sigma(X)$ and $\sigma(Y)$**

*   Show $\sigma(X) \subseteq \sigma(Y)$: Every set in $\sigma(X)$ (e.g., $\{G_1, G_2\}$) is also a member of $\sigma(Y)$ because $\sigma(Y)$ contains every possible subset of $\Omega$.
*   Information: $Y$ contains more information because it allows us to distinguish which machine produced the item. $X$ only tells us if it is good or defective, "blurring" the distinction between Machine 1 and Machine 2.
****

5.   **Marginal Distribution of $X$**

*   $P(X=0) = P(\{G_1, G_2\}) = 0.50 + 0.20 = \mathbf{0.70}$
*   $P(X=1) = P(\{D_1, D_2\}) = 0.10 + 0.20 = \mathbf{0.30}$
****

6.   **Marginal Distribution of $Y$**
*   $P(Y=1) = \mathbf{0.50}$, $P(Y=2) = \mathbf{0.20}$, $P(Y=3) = \mathbf{0.10}$, $P(Y=4) = \mathbf{0.20}$
****

7.   **Joint Probabilities $P(X, Y)$**
*   $P(X=0, Y=1) = P(\{G_1\}) = \mathbf{0.50}$
*   $P(X=0, Y=2) = P(\{G_2\}) = \mathbf{0.20}$
*   $P(X=1, Y=3) = P(\{D_1\}) = \mathbf{0.10}$
*   $P(X=1, Y=4) = P(\{D_2\}) = \mathbf{0.20}$
****

8.   **Conditional Probabilities**
*   Using $P(A|B) = \frac{P(A \cap B)}{P(B)}$:
*   $P(X=1 | Y=3) = \frac{P(X=1, Y=3)}{P(Y=3)} = \frac{0.10}{0.10} = \mathbf{1}$
*   $P(X=1 | Y=4) = \frac{P(X=1, Y=4)}{P(Y=4)} = \frac{0.20}{0.20} = \mathbf{1}$
*   $P(Y=3 | X=1) = \frac{P(Y=3, X=1)}{P(X=1)} = \frac{0.10}{0.30} = \mathbf{1/3}$
*   $P(Y=4 | X=1) = \frac{P(Y=4, X=1)}{P(X=1)} = \frac{0.20}{0.30} = \mathbf{2/3}$   
****

9.   **Interpret the Difference in Conditioning**
*   Conditioning on $Y$ provides "perfect" information about the outcome. There is no uncertainty left regarding $X$.
*   Conditioning on $X$ only narrows the possibilities. If we know $X=1$, we still don't know if the item came from Machine 1 or Machine 2; we only know the probabilities of each.
****

10.   **Information Availability**
*   In $\sigma(X)$ but not in $\sigma(Y)$: None. Since $\sigma(X) \subseteq \sigma(Y)$, all information in $X$ is already contained within $Y$.
*   In $\sigma(Y)$ but not in $\sigma(X)$: The specific Machine ID. $\sigma(Y)$ can resolve events like $\{G_1\}$ (Machine 1 only), whereas $\sigma(X)$ can only resolve the union $\{G_1, G_2\}$.





11.   **Python Visualization**





```python
import matplotlib.pyplot as plt
import numpy as np

# Sample space probabilities
omega = ["G1", "G2", "D1", "D2"]
p = np.array([0.50, 0.20, 0.10, 0.20])

# Random variables
X = {"G1": 0, "G2": 0, "D1": 1, "D2": 1}
Y = {"G1": 1, "G2": 2, "D1": 3, "D2": 4}

# Marginals
px = {
    0: p[0] + p[1],
    1: p[2] + p[3]
}

py = {
    1: p[0],
    2: p[1],
    3: p[2],
    4: p[3]
}



# Joint distribution table for (X,Y)
# Rows correspond to X = 0, 1
# Columns correspond to Y = 1, 2, 3, 4

joint = np.array([
    [0.50, 0.20, 0.00, 0.00],  # X = 0
    [0.00, 0.00, 0.10, 0.20]   # X = 1
])

fig, ax = plt.subplots(figsize=(8, 4))

im = ax.imshow(joint, cmap="YlOrRd", aspect="auto")

# Axis labels
ax.set_xticks(range(4))
ax.set_xticklabels([1, 2, 3, 4])
ax.set_yticks(range(2))
ax.set_yticklabels([0, 1])

ax.set_xlabel("Y")
ax.set_ylabel("X")
ax.set_title("Heat Map of the Joint Distribution of $(X,Y)$")

# Add probability values inside cells
for i in range(joint.shape[0]):
    for j in range(joint.shape[1]):
        ax.text(j, i, f"{joint[i, j]:.2f}",
                ha="center", va="center", color="black", fontsize=12)

# Color bar
cbar = plt.colorbar(im, ax=ax)
cbar.set_label("Probability")

plt.tight_layout()
plt.show()

# Conditionals
p_y3_given_x1 = p[2] / (p[2] + p[3])
p_y4_given_x1 = p[3] / (p[2] + p[3])

fig, axes = plt.subplots(2, 2, figsize=(12, 8))
fig.suptitle("Visualizing Information in Two Random Variables", fontsize=14)

# 1. Sample space probabilities
axes[0, 0].bar(omega, p, color=["steelblue", "steelblue", "salmon", "salmon"])
axes[0, 0].set_title("Sample space probabilities")
axes[0, 0].set_ylabel("Probability")
for i, val in enumerate(p):
    axes[0, 0].text(i, val + 0.02, f"{val:.2f}", ha="center")

# 2. Partition induced by X
axes[0, 1].axis("off")
axes[0, 1].set_title(r"Partition induced by $\sigma(X)$")
axes[0, 1].text(
    0.1, 0.65,
    "{G1, G2}   ->   X = 0  (Good)\n\n{D1, D2}   ->   X = 1  (Defective)",
    fontsize=12,
    bbox=dict(facecolor="lightyellow", edgecolor="black")
)

# 3. Partition induced by Y
axes[1, 0].axis("off")
axes[1, 0].set_title(r"Partition induced by $\sigma(Y)$")
axes[1, 0].text(
    0.1, 0.45,
    "{G1} -> Y = 1\n{G2} -> Y = 2\n{D1} -> Y = 3\n{D2} -> Y = 4",
    fontsize=12,
    bbox=dict(facecolor="lavender", edgecolor="black")
)

# 4. Marginal of X and conditional of Y given X=1
x_labels = ["P(X=0)", "P(X=1)", "P(Y=3 | X=1)", "P(Y=4 | X=1)"]
x_vals = [px[0], px[1], p_y3_given_x1, p_y4_given_x1]
axes[1, 1].bar(x_labels, x_vals, color=["steelblue", "salmon", "orange", "orange"])
axes[1, 1].set_title("Marginal and conditional probabilities")
axes[1, 1].set_ylabel("Probability")
for i, val in enumerate(x_vals):
    axes[1, 1].text(i, val + 0.02, f"{val:.2f}", ha="center")

plt.tight_layout()
plt.show()
```

12.   **Shannon Entropy of $X$ and $Y$**


> Shannon Entropy measures the average level of "information" or uncertainty in a random variable, calculated as $H(V) = -\sum p(v) \log_2 p(v)$.

Marginal Entropy


*   $H(X)$: Based on the distribution $P(X=0)=0.70$ and $P(X=1)=0.30$.$$H(X) = -(0.7 \log_2 0.7 + 0.3 \log_2 0.3) \approx \mathbf{0.8813 \text{ bits}}$$
*   $H(Y)$: Based on the distribution $P(Y)=\{0.5, 0.2, 0.1, 0.2\}$.$$H(Y) = -(0.5 \log_2 0.5 + 0.2 \log_2 0.2 + 0.1 \log_2 0.1 + 0.2 \log_2 0.2) \approx \mathbf{1.7610 \text{ bits}}$$

Joint Entropy


*   $H(X, Y)$: Since $X$ is a deterministic function of $Y$ (knowing the machine and quality automatically tells you the quality), the joint uncertainty is simply the uncertainty of the most detailed variable.$$H(X, Y) = H(Y) \approx \mathbf{1.7610 \text{ bits}}$$

Conditional Entropy


*   $H(X|Y)$: The uncertainty remaining in $X$ when $Y$ is known. Since $Y$ perfectly predicts $X$:$$H(X|Y) = \mathbf{0 \text{ bits}}$$
*   $H(Y|X)$: The uncertainty remaining in $Y$ (which machine was used) when only $X$ (quality) is known.$$H(Y|X) = H(X, Y) - H(X) = 1.7610 - 0.8813 \approx \mathbf{0.8797 \text{ bits}}$$












# **Q10) Information Available in Continuous Random Varaibles**

1.   Probability Space $(\Omega, \mathfrak{F}, P)$

*   Sample Space ($\Omega$): $\mathbb{R}^2$.
*   $\sigma$-algebra ($\mathfrak{F}$): The Borel $\sigma$-algebra on $\mathbb{R}^2$, denoted as $\mathfrak{B}(\mathbb{R}^2)$.
*   Probability Measure ($P$): For any set $A \in \mathfrak{F}$, $P(A) = \iint_A f_{T,P}(t, p) \, dt \, dp$, where $f_{T,P}(t,p) = \frac{1}{4\pi} \exp\left(-\frac{t^2}{2} - \frac{p^2}{8}\right)$.
****

2.   $\sigma$-algebra generated by $X$, $\sigma(X)$

*   $X(t,p) = t$. This random variable only "sees" the temperature.
*   $\sigma(X) = \{ B \times \mathbb{R} : B \in \mathfrak{B}(\mathbb{R}) \}$. It consists of "vertical strips" in the $t-p$ plane.
****

3.   $\sigma$-algebra generated by $Y$, $\sigma(Y)$

*   $Y(t,p) = (t,p)$. This is the identity mapping on $\Omega$.
*   $\sigma(Y) = \mathfrak{B}(\mathbb{R}^2)$, which is the full Borel $\sigma$-algebra on the plane.
****

4.    Show $\sigma(X) \subset \sigma(Y)$

*   Since every set of the form $B \times \mathbb{R}$ is a Borel set in $\mathbb{R}^2$, $\sigma(X) \subseteq \sigma(Y)$.
*   Information Explanation: $Y$ contains more information because it specifies both $T$ and $P$, whereas $X$ only specifies $T$. Knowing $Y$ allows you to determine $X$ exactly, but knowing $X$ leaves $P$ unknown.
****

5.    Marginal Density of $X$

> Since $T \sim N(0, 1)$:
*   $f_X(x) = \frac{1}{\sqrt{2\pi}} \exp\left(-\frac{x^2}{2}\right)$.
****

6.   Marginal Density of $Y$
> $Y$ is the joint pair $(T, P)$. The marginal density of $Y$ is simply the joint density $f_{T,P}(t,p)$:
*   $f_Y(t,p) = \frac{1}{4\pi} \exp\left(-\frac{t^2}{2} - \frac{p^2}{8}\right)$.
****

7.   Probabilities
*   $P(X \leq 0)$: Since $X \sim N(0,1)$, $P(X \leq 0) = \mathbf{0.5}$.
*   $P(X > 1)$: Using Z-tables, $1 - P(Z \leq 1) \approx 1 - 0.8413 = \mathbf{0.1587}$.
*   $P(Y \in (-\infty, 0] \times \mathbb{R})$: This is equivalent to $P(T \leq 0)$, which is $\mathbf{0.5}$.
*   $P(Y \in [-1, 1] \times [-2, 2])$: Due to independence: $P(-1 \leq T \leq 1) \times P(-2 \leq P \leq 2)$.
*   $P(-1 \leq T \leq 1) \approx 0.6826$.
*   For $P \sim N(0, 4)$, the standard deviation is 2. So $P(-2 \leq P \leq 2)$ is the probability within 1 standard deviation $\approx 0.6826$.
*   Total $\approx 0.6826 \times 0.6826 \approx \mathbf{0.466}$.
****

8.   Conditional Distribution of $X$ given $Y = (t, p)$
*   Because $X$ is a direct component of $Y$ ($X=t$), the distribution is a Dirac delta function centered at $t$: $\delta(x - t)$. There is no uncertainty.
****

9.   Conditional Distribution of $Y$ given $X = x$
*   Given $X=x$, the value of $T$ is fixed at $x$. Because $T$ and $P$ are independent, the distribution of $P$ remains $N(0, 4)$.
*   $f_{Y|X}(t, p | x) = \delta(t - x) \cdot \frac{1}{\sqrt{8\pi}} \exp\left(-\frac{p^2}{8}\right)$.
****

10.   Information Conditioning
*   Conditioning on $X$ fixes the temperature but leaves the pressure unknown (governed by its original marginal distribution).
*   Conditioning on $Y$ fixes both temperature and pressure, removing all uncertainty from the system.
****

11.   $X = \pi_1 \circ Y$
*   Since $X(t,p) = t$ and $\pi_1(Y(t,p)) = \pi_1(t,p) = t$, the equality holds.
*   Measurability: Since $\pi_1$ is a continuous projection, it is Borel measurable. A measurable function of a random variable is itself a random variable relative to the $\sigma$-algebra generated by that variable.
*****

13.   Differential Entropy $h(X)$

> For a Gaussian $N(\mu, \sigma^2)$, $h = \frac{1}{2} \ln(2\pi e \sigma^2)$

*   For $X \sim N(0, 1)$: $h(X) = \frac{1}{2} \ln(2\pi e) \approx \mathbf{1.4189 \text{ nats}}$.
****

14.   Differential Entropy $h(Y)$
> For independent variables, $h(T, P) = h(T) + h(P)$.
*   $h(P)$ where $P \sim N(0, 4)$: $\frac{1}{2} \ln(2\pi e \cdot 4) = \frac{1}{2} \ln(8\pi e)$.
*   $h(Y) = \frac{1}{2} \ln(2\pi e) + \frac{1}{2} \ln(8\pi e) = \mathbf{\ln(4\pi e) \approx 3.483 \text{ nats}}$.
****

15.   Conditional Differential Entropy $h(Y|X)$
*   Due to independence, $h(Y|X) = h(P) = \frac{1}{2} \ln(8\pi e) \approx \mathbf{2.064 \text{ nats}}$.
*   Physical Interpretation: This represents the average uncertainty remaining about the full sensor state when the temperature is already known.
****

16.   $h(X|Y)$ and the Continuous Case vs Discrete
*   What happens to $h(X|Y)$: In this model, $h(X|Y) \to -\infty$.
*   Explanation: In the discrete case, if $X$ is determined by $Y$, $H(X|Y) = 0$. However, in the continuous case, as the distribution becomes a delta function (zero uncertainty), the differential entropy goes to negative infinity. This reflects that a perfectly known continuous value requires "infinite" information to specify exactly compared to a continuous background.

### Python Visulaization

#### The $\sigma$-algebra visualizations

Here is a good Python visualization of the two $\sigma$-algebras in $\mathbb R^2$ for the example

$$
X(t,p)=t, \qquad Y(t,p)=(t,p).
$$

The idea is:

* $\sigma(X)$ consists of sets of the form
  $$
  B\times \mathbb R,
  $$
  so visually these are **vertical strips** in the plane;

* $\sigma(Y)=\mathfrak B(\mathbb R^2)$ consists of all Borel sets in the plane, so visually we can show a general **rectangle**, **disk**, or other 2D region.

This code creates a figure with:

1. a vertical strip representing an event in $\sigma(X)$,
2. a rectangle representing an event in $\sigma(Y)$,
3. an overlay comparison showing that $\sigma(X)$ is coarser than $\sigma(Y)$.



```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle, Circle

fig, axes = plt.subplots(1, 3, figsize=(15, 4.8))

xlim = (-4, 4)
ylim = (-8, 8)

# -------------------------
# (1) Event in sigma(X)
# -------------------------
ax = axes[0]
ax.set_title(r"Event in $\sigma(X)$: $B\times\mathbb{R}$")
ax.set_xlim(xlim)
ax.set_ylim(ylim)
ax.set_xlabel(r"$t$")
ax.set_ylabel(r"$p$")

# Example B = [-1,1]
strip = Rectangle((-1, ylim[0]), 2, ylim[1] - ylim[0],
                  facecolor="skyblue", edgecolor="navy", alpha=0.55)
ax.add_patch(strip)

ax.text(0, 0, r"$[-1,1]\times \mathbb{R}$",
        ha="center", va="center", fontsize=12,
        bbox=dict(facecolor="white", edgecolor="none", alpha=0.8))

ax.axhline(0, color="black", lw=0.8)
ax.axvline(0, color="black", lw=0.8)
ax.grid(alpha=0.25)

# -------------------------
# (2) Event in sigma(Y)
# -------------------------
ax = axes[1]
ax.set_title(r"Event in $\sigma(Y)$: general Borel subset")
ax.set_xlim(xlim)
ax.set_ylim(ylim)
ax.set_xlabel(r"$t$")
ax.set_ylabel(r"$p$")

# Example rectangle in sigma(Y)
rect = Rectangle((-1.5, -3), 3, 6,
                 facecolor="lightgreen", edgecolor="darkgreen", alpha=0.55)
ax.add_patch(rect)

# Also add a disk to emphasize generality
disk = Circle((2.2, 3.5), 1.2,
              facecolor="salmon", edgecolor="darkred", alpha=0.45)
ax.add_patch(disk)

ax.text(0, 0, r"$[-1.5,1.5]\times[-3,3]$",
        ha="center", va="center", fontsize=11,
        bbox=dict(facecolor="white", edgecolor="none", alpha=0.8))
ax.text(2.2, 3.5, r"disk",
        ha="center", va="center", fontsize=10,
        bbox=dict(facecolor="white", edgecolor="none", alpha=0.8))

ax.axhline(0, color="black", lw=0.8)
ax.axvline(0, color="black", lw=0.8)
ax.grid(alpha=0.25)

# -------------------------
# (3) Comparison
# -------------------------
ax = axes[2]
ax.set_title(r"Comparison: $\sigma(X)\subseteq\sigma(Y)$")
ax.set_xlim(xlim)
ax.set_ylim(ylim)
ax.set_xlabel(r"$t$")
ax.set_ylabel(r"$p$")

# Vertical strip for sigma(X)
strip2 = Rectangle((-1, ylim[0]), 2, ylim[1] - ylim[0],
                   facecolor="skyblue", edgecolor="navy", alpha=0.35,
                   label=r"$\sigma(X)$ event")
ax.add_patch(strip2)

# Rectangle for sigma(Y)
rect2 = Rectangle((-1.5, -3), 3, 6,
                  facecolor="lightgreen", edgecolor="darkgreen", alpha=0.5,
                  label=r"$\sigma(Y)$ event")
ax.add_patch(rect2)

ax.text(0, 6.5, r"$\sigma(X)$ sees only vertical strips",
        ha="center", fontsize=11,
        bbox=dict(facecolor="white", edgecolor="none", alpha=0.8))
ax.text(0, -6.5, r"$\sigma(Y)$ sees full 2D regions",
        ha="center", fontsize=11,
        bbox=dict(facecolor="white", edgecolor="none", alpha=0.8))

ax.axhline(0, color="black", lw=0.8)
ax.axvline(0, color="black", lw=0.8)
ax.legend(loc="upper right")
ax.grid(alpha=0.25)

plt.tight_layout()
plt.show()
```
